```python


# Data input for total deaths by continent
covid = {
    'continent': ['North America', 'South America', 'Europe', 'Asia', 'Africa', 'Oceania'],
    'Total Deaths': [1046062, 1686828, 1773022, 1405003, 252873, 9336]  # Total deaths for each continent
}

# Create the DataFrame
df = pd.DataFrame(covid)

# Set the index to continent for easier plotting
df.set_index('continent', inplace=True)

# Sort the DataFrame by 'Total Deaths' in ascending order
df_sorted = df.sort_values(by='Total Deaths')

# Create a horizontal bar chart for total deaths
ax = df_sorted.plot(kind='barh', figsize=(10, 6), legend=False)  # Disable legend
plt.title('Total Deaths by Continent')
plt.xlabel('Total Deaths')  # X-axis label
plt.ylabel('Continents')     # Y-axis label
plt.xticks(rotation=0)       # No rotation needed for x-axis labels

# Format x-axis to show whole numbers correctly
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '{:.0f}'.format(x)))




plt.tight_layout()           # Adjust layout to make room for the labels
plt.show()
```
